```python
import numpy as np
```

```python
import dill
rf = dill.load(open("rf2.obj","rb"))


def get_rank(id):
    return id % 13

def get_suit(id):
    return id // 13



def card_to_ind(card):
    return card[0] * 13 + card[1]

def cards_to_1hot(cards):
    cards1h = np.zeros(52)
    for card in cards:
        cards1h[card_to_ind(card)] = 1
    return cards1h
```

```python
all_cards = []
for i in range(52):
    all_cards.append((i // 13, i % 13))

np.random.seed(0)
```

```python
# Add card to known
# Adding high card decreases probability of 2
# Adding low card slightly increases probability of 2

all_res = []
card_to_add1 = (2, 1)
card_to_add2 = (2, 12)
rank_to_check = 10

for i in range(1000):

    all_cards_shuffle = np.random.permutation(all_cards)

    oppPastDiscards = cards_to_1hot(np.vstack((all_cards_shuffle[10:12], card_to_add1)))
    oppKnownCards = cards_to_1hot(np.vstack((all_cards_shuffle[0:3],)))
    ownCards = cards_to_1hot(np.vstack((all_cards_shuffle[3:10],)))
    state = np.hstack((oppPastDiscards, oppKnownCards,  ownCards))
    preds = rf.predict_proba([state])
    prob1 = [pred[0][1] for pred in preds]


    prob_ranks = np.zeros((13,))
    for i in range(len(prob1)):
        prob_ranks[get_rank(i)] += prob1[i]

    prob_ranks /= 4
    with_card = prob_ranks[rank_to_check]

    oppPastDiscards = cards_to_1hot(np.vstack((all_cards_shuffle[10:12], card_to_add2)))
    oppKnownCards = cards_to_1hot(np.vstack((all_cards_shuffle[0:3])))
    ownCards = cards_to_1hot(np.vstack((all_cards_shuffle[3:10],)))
    state = np.hstack((oppPastDiscards, oppKnownCards,  ownCards))
    preds = rf.predict_proba([state])
    prob1 = [pred[0][1] for pred in preds]


    prob_ranks = np.zeros((13,))
    for i in range(len(prob1)):
        prob_ranks[get_rank(i)] += prob1[i]

    prob_ranks /= 4
    without_card = prob_ranks[rank_to_check]

    all_res.append(with_card-without_card)
print(np.median(all_res))


# Add card to own
# Same card in own decreases probability of rank

all_res = []

to_add1 = (0, 5)
to_add2 = (0, 10)
check = 5

for i in range(1000):
    card = (0, 3)

    all_cards_shuffle = np.random.permutation(all_cards)

    oppKnownCards = cards_to_1hot(np.vstack((all_cards_shuffle[0:3])))
    ownCards = cards_to_1hot(np.vstack((all_cards_shuffle[3:10],to_add1)))
    state = np.hstack((oppPastDiscards, oppKnownCards,  ownCards))
    preds = rf.predict_proba([state])
    prob1 = [pred[0][1] for pred in preds]

    prob_ranks = np.zeros((13,))
    for i in range(len(prob1)):
        prob_ranks[get_rank(i)] += prob1[i]

    prob_ranks /= 4
    with_card = prob_ranks[check]


    oppKnownCards = cards_to_1hot(np.vstack((all_cards_shuffle[0:3])))
    ownCards = cards_to_1hot(np.vstack((all_cards_shuffle[3:10],to_add2)))
    state = np.hstack((oppPastDiscards, oppKnownCards,  ownCards))
    preds = rf.predict_proba([state])
    prob1 = [pred[0][1] for pred in preds]


    prob_ranks = np.zeros((13,))
    for i in range(len(prob1)):
        prob_ranks[get_rank(i)] += prob1[i]

    prob_ranks /= 4
    without_card = prob_ranks[check]

    all_res.append(with_card-without_card)
print(np.mean(all_res))

# Add four cards to discared
#

all_res = []

for i in range(1000):
    card = (0, 3)

    all_cards_shuffle = np.random.permutation(all_cards)

    oppPastDiscards = cards_to_1hot(np.vstack( (all_cards_shuffle[11:13],  (0, 2), (1, 2), (2, 2), (3, 2))))
    oppKnownCards = cards_to_1hot(np.vstack((all_cards_shuffle[0:3])))
    ownCards = cards_to_1hot(np.vstack((all_cards_shuffle[3:10],)))
    state = np.hstack((oppPastDiscards, oppKnownCards,  ownCards))
    preds = rf.predict_proba([state])
    prob1 = [pred[0][1] for pred in preds]


    prob_ranks = np.zeros((13,))
    for i in range(len(prob1)):
        prob_ranks[get_rank(i)] += prob1[i]

    prob_ranks /= 4
    with_card = prob_ranks[2]

    oppPastDiscards = cards_to_1hot(np.vstack( (all_cards_shuffle[11:17])))
    oppKnownCards = cards_to_1hot(np.vstack((all_cards_shuffle[0:3])))
    ownCards = cards_to_1hot(np.vstack((all_cards_shuffle[3:10],)))
    state = np.hstack((oppPastDiscards, oppKnownCards,  ownCards))
    preds = rf.predict_proba([state])
    prob1 = [pred[0][1] for pred in preds]


    prob_ranks = np.zeros((13,))
    for i in range(len(prob1)):
        prob_ranks[get_rank(i)] += prob1[i]

    prob_ranks /= 4
    without_card = prob_ranks[2]

    all_res.append(with_card-without_card)
print(np.mean(all_res))
```
